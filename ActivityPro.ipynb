# Tidying the data

```python
X=df.loc[:, df.columns != 'Activity']
y=df.Activity
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 2, shuffle = True, stratify = y)
```

```python
df.median()
```

```python
X_train.head()
```

```python
# replace the nans in the training data with the median
median = X_train.median()
X_train.fillna(value=median, inplace=True)
```

```python
# replace the nans in the test data with the median
medianTest = X_test.median()
X_test.fillna(value=medianTest, inplace=True)
```

```python
#scale 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
scalerTest = StandardScaler()
X_test = scalerTest.fit_transform(X_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA().fit(X=X_train,y=y_train)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
pca1 = PCA(0.999).fit(X=X_train,y=y_train)
pca1.n_components_
```

```python
X_train
```

```python
componentsTrain = pca1.transform(X_train)
componentsTrain.shape
```

```python
componentsTest = pca1.transform(X_test)
componentsTest.shape
```

```python
results={}
```

```python
dfdrop = df.dropna(axis=1)
Xdrop=dfdrop.loc[:, dfdrop.columns != 'Activity']
ydrop=dfdrop.Activity
X_traindrop, X_testdrop, y_traindrop, y_testdrop = train_test_split(Xdrop, ydrop, test_size = 0.33, random_state = 2, shuffle = True, stratify = ydrop)
```

# Classification

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

#nulls replaced with median + pca
decision_tree_classifier = DecisionTreeClassifier()
decision_tree_classifier.fit(componentsTrain, y_train)
results["Tree_median_pca"]=decision_tree_classifier.score(componentsTest, y_test)
```

```python
#nulls replaced with median + no pca
decision_tree_classifier = DecisionTreeClassifier()
decision_tree_classifier.fit(X_train, y_train)
results["Tree_median"]=decision_tree_classifier.score(X_test, y_test)
```

```python
#dropping all the cols with nulls
decision_tree_classifier = DecisionTreeClassifier()
decision_tree_classifier.fit(X_traindrop, y_traindrop)
results["Tree_dropcols"]=decision_tree_classifier.score(X_testdrop, y_testdrop)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_classifier = RandomForestClassifier(criterion="entropy",n_estimators=25)
random_forest_classifier.fit(componentsTrain, y_train)
results["Forest_median_pca"]=random_forest_classifier.score(componentsTest, y_test)
```

```python
random_forest_classifier = RandomForestClassifier(criterion="entropy",n_estimators=25)
random_forest_classifier.fit(X_train, y_train)
results["Forest_median"]=random_forest_classifier.score(X_test, y_test)
```

```python
random_forest_classifier = RandomForestClassifier(criterion="entropy",n_estimators=25)
random_forest_classifier.fit(X_traindrop, y_traindrop)
results["Forest_dropcols"]=random_forest_classifier.score(X_testdrop, y_testdrop)
```

## Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
logclass=LogisticRegression(max_iter=10000)
logclass.fit(componentsTrain, y_train)
results["LogisticRegression_median_pca"]=logclass.score(componentsTest, y_test)
```

```python
logclass=LogisticRegression(max_iter=10000)
logclass.fit(X_train, y_train)
results["LogisticRegression_median"]=logclass.score(X_test, y_test)
```

```python
logclass=LogisticRegression(max_iter=10000)
logclass.fit(X_traindrop, y_traindrop)
results["LogisticRegression_dropcols"]=logclass.score(X_testdrop, y_testdrop)
```

## KNeighbors
